# Chapter 1 Lab: Lookup Table vs Rule Model

## 1. Goal

Observe the difference between replaying stored answers and generating a testable prediction from a relationship.

## 2. Theory

Both approaches can be correct on seen inputs. The key observation is what happens on an unseen input and after a counterexample appears.

```python
known_answers = {1: 3, 2: 5, 3: 7}

def lookup(x: int):
    return known_answers.get(x, 'unknown')

def linear_rule(x: int) -> int:
    return 2 * x + 1

for x in [1, 2, 3, 4]:
    print(x, lookup(x), linear_rule(x))
```

Expected observation: both approaches return the same values for 1, 2, and 3. The lookup table returns `unknown` for 4; the rule predicts 9.

## 3. Experiment: Two Rules Fit the Same Evidence

The second rule adds a term that is zero at every seen input. Predict its output for 4 before running the cell.

```python
def alternative_rule(x: int) -> int:
    return 2 * x + 1 + (x - 1) * (x - 2) * (x - 3)

for x in [1, 2, 3, 4]:
    print(x, linear_rule(x), alternative_rule(x))
```

Expected observation: both rules fit the three seen examples, but they predict 9 and 15 for the unseen input 4. Finite evidence does not uniquely determine the learned rule.

## 4. Experiment: A Counterexample

A new observation is `(4, 10)`. Compute each rule's error. Error identifies a mismatch; it does not decide whether the rule, data, or task definition should change.

```python
observed_x, observed_y = 4, 10

for name, rule in [('linear', linear_rule), ('alternative', alternative_rule)]:
    prediction = rule(observed_x)
    error = observed_y - prediction
    print(name, 'prediction=', prediction, 'error=', error)
```
