# Chapter 11 · Build, Scale, Compare

Predict before running: which coordinate will dominate raw Euclidean distance? How will standardization and task weights change the nearest company?

```python
import numpy as np

names = np.array(['Target', 'LargeStable', 'SmallFast', 'Balanced'])
# Coordinate contract: [revenue_millions, growth_percent, margin_percent]
companies = np.array([
    [500.0, 20.0, 12.0],
    [900.0, 12.0, 18.0],
    [120.0, 28.0, 8.0],
    [480.0, 16.0, 16.0],
])

def euclidean_to_target(matrix):
    return np.linalg.norm(matrix[1:] - matrix[0], axis=1)

raw_distance = euclidean_to_target(companies)
print(list(zip(names[1:], raw_distance)))
```

```python
mean = companies.mean(axis=0)
std = companies.std(axis=0)
standardized = (companies - mean) / std
standardized_distance = euclidean_to_target(standardized)
print(list(zip(names[1:], standardized_distance)))
```

## Task-weighted geometry

Predict what happens when a growth task gives growth three times the weight of other coordinates. Weighting here changes the geometry; it does not reveal a naturally true peer group.

```python
growth_weights = np.array([1.0, 3.0, 1.0])
weighted = standardized * growth_weights
weighted_distance = euclidean_to_target(weighted)
print(list(zip(names[1:], weighted_distance)))
```

```python
def cosine_similarity_to_target(matrix):
    target = matrix[0]
    others = matrix[1:]
    return (others @ target) / (np.linalg.norm(others, axis=1) * np.linalg.norm(target))

cosine = cosine_similarity_to_target(companies)
print(list(zip(names[1:], cosine)))
```
