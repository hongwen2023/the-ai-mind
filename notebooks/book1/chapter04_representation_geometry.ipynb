# Chapter 4 · Representation Geometry Lab

Predict before every run. Inspect what each encoding makes close, distinct, or impossible to recover.

```python
import numpy as np

def ordinal(day: int) -> np.ndarray:
    return np.array([day], dtype=float)

def one_hot(day: int) -> np.ndarray:
    result = np.zeros(7, dtype=float)
    result[day] = 1.0
    return result

def cyclical(day: int) -> np.ndarray:
    angle = 2 * np.pi * day / 7
    return np.array([np.cos(angle), np.sin(angle)])

def distance(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.linalg.norm(a - b))
```

## Experiment 1 · Three weekday geometries

Predict the Monday–Sunday and Monday–Thursday distances for each encoding.

```python
for encode in (ordinal, one_hot, cyclical):
    mon_sun = distance(encode(0), encode(6))
    mon_thu = distance(encode(0), encode(3))
    print(f'{encode.__name__:8s} Mon-Sun={mon_sun:.3f} Mon-Thu={mon_thu:.3f}')
```

## Experiment 2 · Representation collision

Predict what becomes impossible when missing and genuine zero share one encoding.

```python
def unsafe_encode(value):
    return 0.0 if value is None else float(value)

missing = unsafe_encode(None)
genuine_zero = unsafe_encode(0.0)
print({'missing': missing, 'genuine_zero': genuine_zero})
assert missing == genuine_zero
print('No downstream model can separate these states from this representation alone.')
```

## Experiment 3 · Company neighbors depend on scale

Predict whether raw scale or standardized ratios will define the more useful peer group for a margin-focused task.

```python
names = np.array(['LargeLowMargin', 'SmallHighMargin', 'SmallLowMargin', 'MidHighMargin'])
# revenue in millions, gross margin as a ratio
features = np.array([[10000.0, 0.20], [100.0, 0.70], [120.0, 0.20], [1000.0, 0.68]])

def nearest(matrix, index):
    distances = np.linalg.norm(matrix - matrix[index], axis=1)
    distances[index] = np.inf
    return names[np.argmin(distances)]

standardized = (features - features.mean(axis=0)) / features.std(axis=0)
print('Raw neighbor of SmallHighMargin:', nearest(features, 1))
print('Standardized neighbor:', nearest(standardized, 1))
```
