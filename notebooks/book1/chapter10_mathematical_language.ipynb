# Chapter 10 · Translate, Compute, Break

Write meaning before symbols. The goal is not regression; it is auditing a mathematical translation.

## Translation contract

- Object: quarterly advertising spend and sales
- Unit: ten-thousand currency units
- Relation: approximately linear inside the observed range
- Assumption: other major drivers held fixed
- Test: compare prediction with later point-in-time observation
- Boundary: do not extrapolate far beyond observed spend


```python
import numpy as np

w = 1.8
b = 12.0

def predict_sales(ad_spend):
    return w * ad_spend + b

spend = np.array([0.0, 5.0, 10.0, 20.0])
print(np.column_stack([spend, predict_sales(spend)]))
```

## Three fast checks

1. Units: what unit must `w` have?
2. Limiting case: what happens at zero or enormous spend?
3. Perturbation: if spend rises by one unit, what direction and magnitude should sales change?

```python
x = 8.0
epsilon = 0.1
observed_change = predict_sales(x + epsilon) - predict_sales(x)
expected_change = w * epsilon
print('perturbation check:', observed_change, expected_change)
assert np.isclose(observed_change, expected_change)
```

## Break the translation

Keep the arithmetic unchanged but reinterpret `ad_spend` as click count. Explain why the code still runs while the conclusion answers a different question. Then change the input unit from ten-thousand units to single units without changing `w`. Predict the scale error before running.

```python
spend_in_ten_thousands = 10.0
spend_in_single_units = spend_in_ten_thousands * 10_000
correct = predict_sales(spend_in_ten_thousands)
unit_broken = predict_sales(spend_in_single_units)
print('correct scale:', correct)
print('unit-broken scale:', unit_broken)
```
