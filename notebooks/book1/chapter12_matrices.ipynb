# Chapter 12 · Mix, Trace, Compose

Predict each output from the equations before using matrix multiplication. Keep the contract: inputs are `[vocal, guitar, drums]`; outputs are `[left, right]`.

```python
import numpy as np

x = np.array([1.0, 0.8, 0.5])
A = np.array([
    [0.8, 0.5, 0.3],
    [0.8, 0.4, 0.6],
])

y_manual = np.array([
    0.8*x[0] + 0.5*x[1] + 0.3*x[2],
    0.8*x[0] + 0.4*x[1] + 0.6*x[2],
])
y_matrix = A @ x
print('manual:', y_manual)
print('matrix:', y_matrix)
assert np.allclose(y_manual, y_matrix)
```

## Column influence

Before running, predict the output change when only guitar rises by `0.1`. Which column contains the complete answer?

```python
delta_x = np.array([0.0, 0.1, 0.0])
observed_delta = A @ (x + delta_x) - A @ x
column_prediction = 0.1 * A[:, 1]
print('observed:', observed_delta)
print('column prediction:', column_prediction)
assert np.allclose(observed_delta, column_prediction)
```

## Composition order

Use a two-dimensional state. Predict whether scaling then mixing equals mixing then scaling.

```python
state = np.array([1.0, 2.0])
scale = np.array([[2.0, 0.0], [0.0, 0.5]])
mix = np.array([[1.0, 1.0], [0.0, 1.0]])

scale_then_mix = mix @ (scale @ state)
mix_then_scale = scale @ (mix @ state)
print('scale then mix:', scale_then_mix)
print('mix then scale:', mix_then_scale)
assert not np.allclose(scale_then_mix, mix_then_scale)
```
