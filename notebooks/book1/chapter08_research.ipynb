```python
import numpy as np

rng = np.random.default_rng(7)
n = 200
label = rng.integers(0, 2, size=n)
shape = label.copy()
background = label.copy()

def accuracy(prediction, target):
    return np.mean(prediction == target)

def shape_rule(shape, background):
    return shape

def background_rule(shape, background):
    return background
```

```python
conditions = {
    'original': (shape, background),
    'background_swap': (shape, 1 - background),
    'shape_corruption': (rng.integers(0, 2, size=n), background),
}

for name, (condition_shape, condition_background) in conditions.items():
    shape_score = accuracy(shape_rule(condition_shape, condition_background), label)
    background_score = accuracy(background_rule(condition_shape, condition_background), label)
    print(f'{name:20s} shape={shape_score:.3f} background={background_score:.3f}')
```
