# Chapter 7 · Fit, Hold Out, Shift

Predict train, holdout, and shifted behavior before every run.

```python
import numpy as np
rng = np.random.default_rng(7)
def truth(x): return np.sin(1.5 * x)
x_train = np.linspace(-2, 2, 10)
y_train = truth(x_train) + rng.normal(0, 0.08, len(x_train))
x_test = np.linspace(-1.9, 1.9, 60)
y_test = truth(x_test)
x_shift = np.linspace(2.2, 3.5, 30)
y_shift = truth(x_shift)
def mse(a, b): return float(np.mean((a-b)**2))
def fit_eval(degree):
    c = np.polyfit(x_train, y_train, degree)
    return mse(np.polyval(c,x_train),y_train), mse(np.polyval(c,x_test),y_test), mse(np.polyval(c,x_shift),y_shift)
```

## Experiment 1 · Degree and evidence regions

Predict which degree underfits, interpolates, or becomes unstable outside coverage.

```python
for d in (0, 1, 3, 7, 9):
    train, test, shift = fit_eval(d)
    print(f'degree={d} train={train:.4f} holdout={test:.4f} shift={shift:.4f}')
```

## Experiment 2 · Re-sampling uncertainty

Predict whether one split is enough evidence.

```python
scores=[]
for seed in range(10):
    local=np.random.default_rng(seed)
    ys=truth(x_train)+local.normal(0,0.15,len(x_train))
    c=np.polyfit(x_train,ys,7)
    scores.append(mse(np.polyval(c,x_test),y_test))
print('degree-7 holdout mean=',round(float(np.mean(scores)),4),'std=',round(float(np.std(scores)),4))
```
