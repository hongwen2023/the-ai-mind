# Chapter 6 · Feedback-to-Update Lab

Predict parameter direction and future behavior before every run. No Autograd is used.

```python
def predict(theta, x): return theta * x

def update(theta, x, target, rate):
    prediction = predict(theta, x)
    error = prediction - target
    new_theta = theta - rate * error * x
    return new_theta, {'x': x, 'target': target, 'prediction': prediction, 'error': error, 'theta_before': theta, 'theta_after': new_theta}

def train(samples, rate, epochs=5, reset_each_sample=False):
    theta, trace = 0.0, []
    for _ in range(epochs):
        for x, target in samples:
            if reset_each_sample: theta = 0.0
            theta, entry = update(theta, x, target, rate)
            trace.append(entry)
    return theta, trace
```

## Experiment 1 · Update rate

Predict zero-rate, useful-rate, and excessive-rate behavior.

```python
samples = [(1.0, 2.0), (2.0, 4.0)]
for rate in (0.0, 0.1, 1.5):
    theta, trace = train(samples, rate=rate, epochs=5)
    print('rate=', rate, 'theta=', theta, 'prediction x=3:', predict(theta, 3.0))
```

## Experiment 2 · Wrong and misassigned feedback

Predict which relation each learner will retain.

```python
correct = [(1.0, 2.0), (2.0, 4.0)]
wrong = [(1.0, -2.0), (2.0, -4.0)]
misassigned = [(1.0, 4.0), (2.0, 2.0)]
for name, data in [('correct', correct), ('wrong', wrong), ('misassigned', misassigned)]:
    theta, _ = train(data, rate=0.1, epochs=20)
    print(name, 'theta=', round(theta, 4), 'prediction x=3:', round(predict(theta, 3.0), 4))
```

## Experiment 3 · Retention and coverage

Predict the effect of resetting the parameter and training on one point only.

```python
retained, _ = train(samples, rate=0.1, epochs=10)
reset, _ = train(samples, rate=0.1, epochs=10, reset_each_sample=True)
single, _ = train([(1.0, 2.0)], rate=0.1, epochs=10)
print('retained theta:', retained)
print('reset-each-sample theta:', reset)
print('single-point theta:', single, '(future performance still needs testing)')
```
