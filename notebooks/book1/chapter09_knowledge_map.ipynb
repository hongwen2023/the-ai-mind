## Edge contract

Every edge needs a source, target, claim, boundary, and evidence. Edit the starter graph only when you can explain all five fields.

```python
nodes = [
    'relationship', 'generation', 'abstraction', 'representation',
    'computation', 'learning', 'generalization', 'research_evidence',
]

edges = [
    {'source': 'relationship', 'target': 'generation', 'claim': 'local relations can generate larger patterns', 'boundary': 'iterated interacting systems', 'evidence': 'rule perturbation'},
    {'source': 'generation', 'target': 'abstraction', 'claim': 'generated complexity creates a need for selective reasoning', 'boundary': 'task-specific reasoning', 'evidence': 'successful compression and failure cases'},
    {'source': 'abstraction', 'target': 'representation', 'claim': 'preserved relations need an operable form', 'boundary': 'declared task and interface', 'evidence': 'representation audit'},
    {'source': 'representation', 'target': 'computation', 'claim': 'available encoding constrains valid operations', 'boundary': 'chosen encoding', 'evidence': 'collision and geometry tests'},
    {'source': 'computation', 'target': 'learning', 'claim': 'feedback can change future computation', 'boundary': 'retained update mechanism', 'evidence': 'learning trace'},
    {'source': 'learning', 'target': 'generalization', 'claim': 'updated relations need independent tests', 'boundary': 'stated holdout and shift', 'evidence': 'generalization audit'},
    {'source': 'generalization', 'target': 'research_evidence', 'claim': 'success and failure admit competing explanations', 'boundary': 'declared hypotheses', 'evidence': 'discriminating intervention'},
]
```

```python
def downstream(start, edge_list):
    adjacency = {}
    for edge in edge_list:
        adjacency.setdefault(edge['source'], []).append(edge['target'])
    seen, frontier = set(), [start]
    while frontier:
        current = frontier.pop(0)
        for target in adjacency.get(current, []):
            if target not in seen:
                seen.add(target)
                frontier.append(target)
    return sorted(seen)

print('Downstream of representation:', downstream('representation', edges))
```
